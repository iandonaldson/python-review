```python
# find duplicates in a list of numbers

def find_duplicates(elements):
    duplicates, seen = set(), set()
    for element in elements:
        if element in seen:
            duplicates.add(element)
        seen.add(element)
    return list(duplicates)

find_duplicates([1,2,3,4,4,4])
```

```python
# find pairs of integers in list so that their sum is equal to integer x

def find_pairs(l,x):
    pairs = []
    for a in l:
        for b in l:
            # print(f"a + b is {a} + {b} = {a+b}")
            if a + b == x:
                pairs.append([a,b])
    return(pairs)


l = [1,2,3,4,5]
pairs = find_pairs(l, 9)

pairs
```

```python
# check if a string is a palindrome

this_string = "backcab"
print(type(this_string))

def is_palindrome(s):
    r = list(s)
    r.reverse() # or use reverse(r)
    if list(s) == r:
        return True
    else:
        return False

# simpler    
def is_palindrome(s):
    return s == s[::-1]
    
print(is_palindrome(this_string))
```

```python
# recursion

def exponential(n):
    # base case
    if n == 1:
        return 1

    # recursive case
    return n * exponential(n-1)

answer = exponential(5)

answer
```

```python
# queue

# FIFO
# use for printer jobs and being polite

queue = []

# Add items to the queue
queue.append("Alice")
queue.append("Bob")
queue.append("Charlie")

print("Queue after additions:", queue)

# Remove the first item
removed_item = queue.pop(0)

print("Removed item:", removed_item)
print("Remaining queue:", queue)

# can also be implemented in reverse using insert(0, item) and pop()
# both methods are inefficent pop(0) and insert(0,) are O(n) time

# instead use dequeue

from collections import deque

queue = deque()

queue.append("Alice")
queue.append("Bob")
queue.append("Charlie")

print(f"using dequeue, pop(left) will be: {queue.popleft()}")
```

```python
# reverse string using recursion

# pop?

def reverse(s):
    # base case
    if len(s) == 1:
        return s
    
    # recursive case
    return s[-1] + reverse(s[0:len(s)-1])

s = "elvis"
answer = reverse(s)
print(f"the reverse of {s} is {answer}")
```

```python
# fibonnaci

def fibonnaci(n):
    a = 0
    b = 1
    f = []
    for i in range(n):
        f.append(b)
        a,b = b, a+b
    return f

fibonnaci(10)
```
