Find the first unique integer in an array of integers

```python
def solution(A):
    
    # Check if A is None or empty, and if all elements in A are integers
    if A is None or len(A) == 0:
        return -1
    if not all(isinstance(a, int) for a in A):
        return -1
        
    counts = {}

    # Count the occurrences of each element in A using a dictionary
    for a in A:
        counts[a] = counts.get(a,0) + 1

    for x in A:
        if counts[x] == 1:
            return x

    # More verbose version of the same code
    # if x in counts:
    #     counts[x] += 1
    # else:
    #     counts[x] = 1
    

    return -1
```

Find the first unique integer in an array of integers

```python
A = [9, 3, 9, 3, 9, 7, 9]

answer = solution(A)
print(answer)
```
